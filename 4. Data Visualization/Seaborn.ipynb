```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

__Line plot__

```python
x1 = np.arange(1,11)
y1 = np.log(x1)
plt.figure(figsize=(5, 3.5))
sns.lineplot(x=x1, y=y1)
plt.show()
```

```python
flights = sns.load_dataset("flights")
flights.head()
```

```python
sns.lineplot(data=flights, x="year", y="passengers")
```

```python
sns.lineplot(data=flights, x="year", y="passengers", hue="month")
```

```python
penguins = sns.load_dataset("penguins")
penguins.head()
```

```python
plt.figure(figsize=(8, 3.5))
sns.lineplot(data=penguins, x="bill_length_mm", y="flipper_length_mm")
plt.show()
```

__Countplot__

```python
plt.figure(figsize=(5, 3.5))
sns.countplot(data=penguins, x="species")
plt.show()
```

```python
plt.figure(figsize=(5, 3.5))
sns.countplot(data=penguins, x="species", hue="sex")
plt.show()
```

__Barplot__

```python
plt.figure(figsize=(5, 3.5))
sns.barplot(data=penguins, x="species", y="bill_length_mm")
plt.show()
```

```python
sns.barplot(data=penguins, x="island", y="body_mass_g", hue="sex")
```

__Boxplot__

```python
sns.boxplot(data=penguins, x="species", y="bill_length_mm")
```

```python
sns.boxplot(data=penguins, x="species", y="body_mass_g", showmeans=True)
```

__Strip plot__

```python
plt.figure(figsize=(5, 3.5))
sns.stripplot(data=penguins, x="body_mass_g", y="species")
plt.show()
```

__Cat Plot__

```python
sns.catplot(data=penguins, x="bill_length_mm", y="species")
```

__Hist plot__

```python
sns.histplot(data=penguins, x="bill_length_mm")
```

__Scatter plot__

```python
sns.scatterplot(data=penguins, x="bill_length_mm", y="flipper_length_mm")
```

```python
sns.pairplot(penguins, hue="species")
```

```python
sns.pairplot(penguins, kind="kde")
```

### Reg plot

```python
sns.regplot(data = penguins, x = "bill_length_mm", y = "flipper_length_mm")
```

```python
sns.lmplot(data = penguins, x="bill_length_mm", y="flipper_length_mm", hue="species")
```

__Heatmap__

```python
df = pd.DataFrame(np.random.randn(50).reshape(10,5))
corr = df.corr()
ax1 = sns.heatmap(corr, linewidths=2, vmax=1, vmin=0, square=True, annot=True)
plt.show()
```

```python
df_flights = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/flights.csv')
dfp_flights = df_flights.pivot(index='month', columns='year', values='passengers')
```

```python
plt.figure(figsize=(7,7))
sns.heatmap(dfp_flights, annot=True, fmt='d')
```
