```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(5)
plt.figure(figsize=[5,3.5])
plt.plot(x, x**2);
```

```python
plt.figure(figsize=[5,3.5])
x = np.arange(10)
plt.plot(x, x**1.7, 'or');
```

```python
flights = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/flights.csv')
flights.head()
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv')
df.head()
```

```python
x = flights['year'].unique()
y = flights.groupby(['year'])['passengers'].sum()
plt.figure(figsize=[5,3.5])
plt.plot(x, y);
```

__Line plot__

```python
plt.figure(figsize=[5,3.5])

days = np.arange(1,11)
temperature = np.random.uniform(25,37, size=10)

plt.plot(days,temperature, 'o-')
plt.title("Temperature Record")
plt.xlabel("Day")
plt.ylabel("Temp.")
plt.show()
```

```python
#Set Axis
plt.figure(figsize=[5,3.5])
days = np.arange(1,11)
temp = np.random.randint(20,35, size=10)

plt.plot(days,temp)
plt.legend(['Temp. line'], loc=2)
plt.axis([0,15, 15,37])   #min,max,min,max
plt.show()
```

```python
days = np.arange(1,11)
city1_temp = np.random.randint(20,30, size=10)
city2_temp = np.random.randint(27,37, size=10)

plt.figure(figsize=[6.4,3.5])
plt.plot(days,city1_temp, 'x-g')
plt.plot(days,city2_temp, 'x:y')
plt.xlabel("Day")
plt.ylabel("Temp.")
plt.legend(["City 1","City 2"])
plt.show()
```

__Subplot__

```python
plt.figure(figsize=[9,3.5])
x = np.arange(5)
plt.subplot(1, 2, 1)
plt.plot(x, x**2)
plt.subplot(1, 2, 2)
plt.plot(x, -1*x**2)
plt.show()
```

```python
fig, axes = plt.subplots()
x = np.arange(1,11)
axes.plot(x, x + 0)
axes.plot(x, x + 1, linestyle='--')
axes.plot(x, x + 2, linestyle='-.')
axes.plot(x, x + 3, linestyle=':')
plt.show()
```

__Plotting Time Series__

```python
plt.figure(figsize=[6.4,3.5])
dates = np.arange(np.datetime64('2022-01-01'), np.datetime64('2022-01-11'))
y = np.random.randint(20,35, size=10)
plt.plot(dates, y)
plt.xticks(rotation=30, ha='right')
plt.show()
```

__Bar chart__

```python
plt.figure(figsize=[10,3.5])
x = ["A", "B", "C", "D"]
y = [3, 8, 1, 10]

plt.subplot(1, 2, 1)
plt.bar(x,y)

plt.subplot(1, 2, 2)
plt.bar(x,y)
plt.plot(x,y,'r');
```

```python
plt.figure(figsize=[5,3.5])
speices = df["species"].unique()
bill_length = df.groupby('species')['bill_length_mm'].mean()
bill_depth = df.groupby('species')['bill_depth_mm'].mean()
plt.bar(speices, bill_length)
plt.bar(speices, bill_depth)
plt.legend(['bill_length (mm)', 'bill_depth (mm)'])
plt.show()
```

```python
plt.figure(figsize=[8,3.5])
days = np.arange(1,11)
city1_temp = np.random.randint(20,30, size=10)
city2_temp = np.random.randint(27,37, size=10)
width = 0.4
plt.bar(days,city1_temp, width, color="b", label='city1')
plt.bar(days+width,city2_temp, width, color="g", label='city2')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=[6.4,3.5])
days = np.arange(1,11)
city1_temp = np.random.randint(10,30, size=10)
city2_temp = np.random.randint(20,30, size=10)
city3_temp = np.random.randint(30,37, size=10)
width = 0.25
plt.bar(days-width,city1_temp, width, color="b", label='city1')
plt.bar(days,city2_temp, width, color="g", label='city2')
plt.bar(days+width,city3_temp, width, color="orange", label='city3')
plt.legend()
plt.show()
```

__Histogram__

```python
plt.figure(figsize=[5,3.5])
city1_temp = np.random.randint(25,35, size=15)
plt.hist(city1_temp)
plt.show()
```

```python
plt.figure(figsize=[5,3.5])
plt.hist(city1_temp, bins=5, edgecolor='black', rwidth=0.9)
plt.show()
```

```python
p1 = df.groupby('species').get_group('Adelie')['body_mass_g']
p2 = df.groupby('species').get_group('Chinstrap')['body_mass_g']
p3 = df.groupby('species').get_group('Gentoo')['body_mass_g']

fig, ax = plt.subplots(1, 3, figsize=(10,3.5))
ax[0].hist([p1])
ax[1].hist([p2])
ax[2].hist([p3])

plt.tight_layout()
plt.show()
```

__Pie Charts__

```python
y = np.array([35, 25, 25, 15])
mylabels = ["class-A", "class-B", "class-C", "class-D"]

plt.pie(y, labels = mylabels)
plt.show() 
```

```python
speices = df["species"].unique()
body_mass = df.groupby('species')['body_mass_g'].mean()
y = np.array([35, 25, 25])

myexplode = [0.1, 0, 0]

plt.pie(body_mass, labels = speices, explode = myexplode, autopct='%1.1f%%')
plt.show() 
```

```python
passengers = flights.groupby(['month'])['passengers'].mean()
month = flights['month'].unique()
plt.pie(passengers, labels = month, autopct='%1.1f%%')
plt.show() 
```

__Stack Plot__

```python
days = np.arange(1,11)
city1_temp = np.random.randint(0,15, size=10)
city2_temp = np.random.randint(20,33, size=10)
city3_temp = np.random.randint(25,40, size=10)
labels = ['city1','city2','city3']

plt.stackplot(days, city1_temp, city2_temp, city3_temp, labels=labels)
plt.legend()
plt.show()
```
